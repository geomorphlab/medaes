# Packages

```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
plt.rcParams['figure.dpi'] = 150
```

# Functions and constants

```python
def fluxdirectionality(dp,rdp):
    return rdp/dp

def duneaspect(h,w):
    return h/w
```

```python
m2Pyr_to_mm2Ps = 1e+6/(60*60*24*365)
```

# Read data

```python
data = pd.read_csv('TableS2.csv')
```

```python
h   = data['Large Height (m)']
w   = data['Large Width (m)']
rdp = data['Mean Resultant Flux Magnitude (m^2/yr)']
dp  = data['Mean of Flux Magnitudes (m^2/yr)']
```

# Analysis

```python
fd = fluxdirectionality(dp,rdp)
da = duneaspect(h,w)
```

# Plots

```python
ss = 10

x = dp*m2Pyr_to_mm2Ps

s1 = plt.scatter(fd,da,c=x,vmin=0,vmax=1.4,s=ss)

plt.yscale('log')
plt.xlim(0,1)
plt.ylim(1e-4,1e-1)

plt.colorbar(label='Drift Potential $\\overline{|\\vec{q}|}$ (mm$^2$/s)')
plt.xlabel('Flux Directionality $\\frac{|\\sum\\vec{q}|}{\\sum|\\vec{q}|}$ (-)')
plt.ylabel('Dune Aspect $H/W$ (-)')
```
